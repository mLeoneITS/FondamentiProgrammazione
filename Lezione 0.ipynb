```python
#Ad esempio, partendo dalla stessa lista:
nums = [5, 2, 9, 1]
```

```python
#Potrei avere bisogno di una nuova lista ordinata come nell'esempio precedente:
sorted_nums = sorted(nums)
print(sorted_nums)
```

```python
#Ma potrei avere bisogno di aggiornare la lista esistente
nums = [5, 2, 9, 1]
print('lista originale prima: ', nums)
nums = sorted(nums)
print('lista originale dopo: ', nums)
```

```python
#Oppure di stampare sia la lista originale che la nuova lista ordinata
nums = [5, 2, 9, 1]
print('lista originale: ', nums)
sorted_nums = sorted(nums)
print('lista ordinata: ', sorted_nums)
```

```python
#Oppure di un metodo (ne parleremo più avanti)
#e quindi non voglio che stampi a video il risultato,
#ma che restituisca la lista al chiamante (anche qui, ne parleremo più avanti)

#metodo
def sort_numbers(nums):
  sorted_nums = sorted(nums)
  return sorted_nums

#codice principale che richiama il metodo
nums = [5, 2, 9, 1]
result = sort_numbers(nums);

print('risultato finale: ',result);
```

## 2. Perché imparare a programmare prima
- **Capire la logica**: algoritmi, strutture dati, flusso di controllo.
- **Valutare la qualità**: bug, performance, sicurezza.
- **Essere indipendenti**: senza conoscenza, si diventa 'copiatori' e non sviluppatori.

**ESEMPIO DI CODICE ERRATO** cosa fa (o meglio, dovrebbe fare) questo codice? Perchè si rompe?

```python
# Esempio di codice con bug nascosto
def somma_lista(lista):
    s = 0
    for i in range(len(lista)+1):
        s += lista[i]
    return s

# Prova con una lista
try:
    print(somma_lista([1, 2, 3]))
except Exception as e:
    print('Errore:', e)
```

## 3. L'AI non è il male, ma...
- **NON È IL MALE**: aiuta a scrivere codice più veloce, supporta debugging e documentazione.
- **PUÒ ESSERE UN MALE SE...** si usa senza capire cosa fa, si delega tutto → perdita di competenze, si ignora la sicurezza.

**CONSAPEVOLEZZA > DIPENDENZA**.

## 4. Come usarla bene
- **DOPO AVER IMPARATO LE BASI**: usala per accelerare, non per sostituire il pensiero critico.
- **BEST PRACTICE**: verifica sempre il codice, usa l'AI come assistente, non come pilota.

### Esempio di Prompt

**Scrivi un programma in Python che ordini una lista di numeri in ordine crescente e spiega ogni passaggio con commenti chiari.**

Riportiamo questa in una chat con l'AI...

```python
# Programma per ordinare una lista di numeri in ordine crescente

# 1. Definiamo una lista di numeri non ordinati
numeri = [5, 2, 9, 1, 7]

# 2. Stampiamo la lista originale per confronto
print("Lista originale:", numeri)

# 3. Usiamo il metodo sort() per ordinare la lista in modo crescente
# Il metodo sort() modifica la lista originale
numeri.sort()

# 4. Stampiamo la lista ordinata
print("Lista ordinata (crescente):", numeri)

# --- Variante alternativa ---
# Se vogliamo mantenere la lista originale e creare una nuova lista ordinata,
# possiamo usare la funzione sorted(), che restituisce una nuova lista ordinata.
numeri_originali = [5, 2, 9, 1, 7]
lista_ordinata = sorted(numeri_originali)

print("Lista originale (immutata):", numeri_originali)
print("Nuova lista ordinata:", lista_ordinata)

```
